##선택정렬(Selection Sort)
- **O(n<sup>2)**

```
1. 입력값에서 최솟값 찾기
2. 정렬된 리스트에 최솟값을 삽입
```

````
%%time
# ex)
입력값 = [5, 10 ,66, 77, 32 , 11, 15]
정렬된리스트 = []

while 입력값:
  정렬된리스트.append(min(입력값))
  del 입력값[입력값.index(min(입력값))]
  print(정렬된리스트)
  print(입력값, '----------------', sep = '\n')

print('----- 최종 -----')
print(정렬된리스트)
print(입력값, '----------------', sep = '\n')
````

##삽입정렬(Insertion Sort)
- **O(n<sup>2)**

```
1. 입력값에서 값을 빼냄
2. 정렬된리스트에서 값을 비교 후 맞는 위치에 삽입
```

````
%%time
# ex)
입력값 = [5, 10 ,66, 77, 32 , 11, 15]
정렬된리스트 = []

def sSort(정렬된리스트, 삽입값):
  for i in range(len(정렬된리스트)):
    if 삽입값 < 정렬된리스트[i]:
      return i
  return len(정렬된리스트)

while 입력값:
  삽입값 = 입력값.pop(0)
  정렬된리스트.insert(sSort(정렬된리스트, 삽입값), 삽입값)
    
  print(정렬된리스트)
  print(입력값, '----------------', sep = '\n')

print('----- 최종 -----')
print(정렬된리스트)
print(입력값, '----------------', sep = '\n')
````

##병합정렬(Merge Sort)
- **O(nlogn)**

```
1. 분할
2. 정복
```

````
%%time
# ex)
lst = [5, 10 ,66, 77, 52, 32 , 11, 15]
def merge(left, right):
  result = []
  while left and right:
    if left[0] < right[0]:
      result.append(left.pop(0))
    else:
      result.append(right.pop(0))
  while left:
    result.append(left.pop(0))
  while right:
    result.append(right.pop(0))
  return result

def mergeSort(lst):
  if len(lst) <= 1:
    return lst
  mid = len(lst) // 2
  left = mergeSort(lst[:mid])
  right = mergeSort(lst[mid:])
  return merge(left, right)

print(mergeSort(lst))
````

##퀵정렬(Quick Sort)
- **O(nlogn)**
- **worst case: O(n<sup>2)**

```
1. 피봇을 이용한 분할
2. 정복
```

````
%%time
# ex)
lst = [5, 10 ,66, 77, 52, 32 , 11, 15]

def quickSort(lst):
  if len(lst) <= 1:
    return lst
  print(lst)
  pivot = lst.pop(0)
  less = [elem for elem in lst if elem < pivot]
  greater = [elem for elem in lst if elem >= pivot]
  tList = less + [pivot] + greater
  print(f'{less} + {[pivot]} + {greater}')
  print(tList)
  print('------------------------------------')
  return quickSort(less) + [pivot] + quickSort(greater)

print(quickSort(lst))
````
