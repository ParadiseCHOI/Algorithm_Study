##class 규칙
1. class name 첫 문자가 대문자
2. class는 대부분 직접만지지 않는다
3. class는 틀 instance가 실제 제품


```
class Car(object):
  maxSpeed = 300
  maxPeople = 5
  def start(self):
    print('출발')
  def stop(self):
    print('정지')
  def __str__(self): # magic method
    return 'hello world'
  def __init__(self):
    print('인스턴스 생성')

class Hybrid(Car):
  battery = 1000
  batteryKm = 300

k9 = Car()
k5 = Hybrid()
print(k9.maxPeople)
print(k5.maxPeople)
print(type(k9))
print(type(k5))
print(dir(k9))
print(dir(k5))
print(k9)
```
