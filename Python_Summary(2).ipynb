##리스트

```
# 변경이 가능한 자료형
# 순서가 있는 자료형
l = [100, 200 ,300, 400]
print(l[::-1])
# l[start:stop:step]
print(l)
print(type(l))
l[1] = 1000
print(l)
```

```
l = [100, 200 ,300, 400]
print(dir(l))
l.append(300)
print(l)
# l.clear()
# l.copy()
print(l.count(300))
l.extend([100, 200 ,300])
print(l)
print(l.index(100))
l.insert(3, 1000)
print(l)
l.pop()
print(l)
l.remove(100)
print(l)
l.reverse()
print(l)
l.sort()
print(l)
# sorted() - return값만 만짐
# reversed() - return값만 만짐
```

##튜플

```
# 변경이 불가능한 자료형 - 참조 불변, 값이 불변은 아님
# 순서가 있는 자료형
l = [10, 20]
t = (l, 200 ,300)
print(t)
print(type(t))
print(dir(t))
l[0] = 10000
print(t)
```

##set

```
# 순서가 없고
# 값의 중복을 허락하지 않는 자료형
s = {100, 200, 300, 300, 300}
print(s)
print(type(s))
print(dir(s))

s.add(500)
print(s)
a = 'aaabbcccddefgggg'
b = set(a)
print(b)
a = list(b)
print(a)
```

##dictionary(class 'dict')


```
# 순서가 없고
# 키의 중복을 허락하지 않는 자료형
d = {'one':10, 'two':20}
print(d['one'])
print(type(d))
print(dir(d))
print(d.values())
print(d.keys())
print(d.items())
l = list(d.items())
print(l[0][1])
```

```
jeju = {'banana':5000, 'orange':2000}
seoul = jeju.copy()
jeju['orange'] = 10000
print(seoul)
print(jeju)
```
