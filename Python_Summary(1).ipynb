##연산자

###산술연산

```
a = 3
b = 10
print(a + b)
print(a - b)
print(b / a)  #float
print(b // a) #int
print(b * a)
print(b ** a)
print(b % a)
```

###비교연산

```
a = 10
b = 3
print(a >= b)
print(a > b)
print(a < b)
print(a <= b)
print(a == b)
print(a != b)
```

###논리연산

```
a = True
b = False
print(a and b)
print(a or b)
print(not b)
```

###할당연산

```
a = 10
a = a + 10
a += 10
print(a)
```

###비트연산

```
a = 40
b = 14
print(bin(a)[2:].zfill(6))
print(bin(b)[2:].zfill(6))
print(bin(a & b)[2:].zfill(6))
print(a & b)
```

##함수
1. 반복문의 경우 Bottom-up(작은 문제에서 출발)
2. 재귀함수의 경우 Top-down(큰 문제에서 출발)

```
def f(x ,y):
  z = x + y
  return z
print(f(3, 5))
```

```
def f():
  print('1')
  print('2')
print('3')
print(4)
f()
```

```
a = 10
def aplus(param):
  param += 10
  return param
print(aplus(a))
print(a)
```

```
a = 10
def aplus():
  global a
  a += 10
  return a
print(aplus())
print(a)
```
