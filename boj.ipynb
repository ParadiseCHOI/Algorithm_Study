##**DP - 10844번** 쉬운 계단 수

```
import sys

val = int(input())
# count = int(sys.stdin.readline()))
mod = 1000000000

dp = [[0 for i in range(10)] for j in range(val+1)]

for i in range(1, 10):
  dp[1][i] = 1

for i in range(2, val+1):
  for j in range(10):
    if j == 0:
      dp[i][j] = dp[i-1][j+1]
    elif j == 9:
      dp[i][j] = dp[i-1][j-1]
    else:
      dp[i][j] = dp[i-1][j-1] + dp[i-1][j+1]

print(sum(dp[val])%mod)

```

#**그래프**

##**그래프 - 1260번** DFS와 BFS

```
import sys

f = open("input.txt", 'r')
V, E, s = map(int, f.readline().split())

graph = [[0] * (V+1) for _ in range(V+1)]
visited = [False] * (V+1)

for _ in range(E):
    i, j = map(int, f.readline().split())
    graph[i][j] = graph[j][i] = 1

# DFS(Depth-First Search)
# > 재귀를 통해 최대 깊이까지 순차 방문
def dfs(s):
    print(s, end=' ')
    visited[s] = True
    for i in range(1, V+1):
        if (not visited[i]) and (graph[s][i] == 1):
            dfs(i)

# BFS(Breadth-First Search)
# > 큐에 가능한 모든 정점을 넣어둔 후 큐에서 정점을 꺼내가며 방문
def bfs(s):
    queue = [s]
    visited[s] = False
    while queue:
        s = queue.pop(0)
        print(s, end=' ')
        for i in range(1, V+1):
            if (visited[i]) and (graph[s][i] == 1):
                queue.append(i)
                visited[i] = False

dfs(s)
print()
bfs(s)
```

##**그래프 - 11724번** 연결 요소의 개수

```
import sys
sys.setrecursionlimit(10000)

f = open("input.txt", 'r')
V, E = map(int, f.readline().split())

graph = [[0] * (V+1) for _ in range(V+1)]
visited = [False] * (V+1)
count = 0

for _ in range(E):
    i, j = map(int, f.readline().split())
    graph[i][j] = 1

# DFS(Depth-First Search)
# > 재귀를 통해 최대 깊이까지 순차 방문
def dfs(s):
    visited[s] = True
    for i in range(1, V+1):
        if (not visited[i]) and (graph[s][i] == 1):
            dfs(i)

def dfsAll():
    global count
    for i in range(1, V+1):
        if not visited[i]:
            dfs(i)
            count += 1

dfsAll()
print(count)
```

##**그래프 - 1707번** 이분그래프

##**이분 탐색 - 11662번** 민호와 강호

```
# 딕셔너리
import sys
import math
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

def getDistance(Ax, Ay, Cx, Cy):
    return math.sqrt(math.pow(Cx-Ax, 2) + math.pow(Cy-Ay, 2))


# 입력
# 민호: A(Ax, Ay) => B(Bx, By)
# 강호: C(Cx, Cy) => D(Dx, Dy)
# Ax, Ay, Bx, By, Cx, Cy, Dx, Dy
# 0 <= 입력값 <= 10000
Ax, Ay, Bx, By, Cx, Cy, Dx, Dy = map(int, f.readline().split())
s = 1000000

Adx = (Bx-Ax) / s
Ady = (By-Ay) / s
Cdx = (Dx-Cx) / s
Cdy = (Dy-Cy) / s

shortest = getDistance(Ax, Ay, Cx, Cy)
print(shortest)
i = 1
while i <= s:
    tmp = getDistance(Ax + Adx*i, Ay + Ady*i, Cx + Cdx*i, Cy + Cdy*i)
    if shortest > tmp:
        shortest = tmp
    i+=1
        
# 출력 
# 민호와 강호가 가장 가까웠을 때의 거리
print(shortest)
```

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

def divide(x, y, size):
    isSameNum = True

    for i in range(x, x + size):
        for j in range(y, y + size):
            if paper[x][y] != paper[i][j]:
                isSameNum = False
                break

    if not isSameNum:
        for i in range(x, x+size, size//3):
            for j in range(y, y+size, size//3):
                divide(i, j, size//3)
    else:
        ans[paper[x][y]] += 1
        return
    

# 입력
# N: N×N크기의 행렬로 표현되는 종이
# N은 3^k 꼴
N = int(f.readline())
paper = [list(map(int, f.readline().split())) for _ in range(N)]
ans= {i:0 for i in range(-1, 2)}

divide(0, 0, len(paper))

# 출력 
# 첫째 줄: -1로만 채워진 종이의 개수
# 둘째 줄: 0으로만 채워진 종이의 개수
# 셋째 줄: 1로만 채워진 종이의 개수
print(*ans.values(), sep='\n')
```

##**분할 정복 - 11729번** 하노이 탑 이동 순서

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

# N: 원판의 수
# source => destination
# other: 나머지 장대

def hanoi(N, source, destination, other):
    if N < 2:
        ans.append([source, destination])
        return

    # N-1개의 원판을 2번 장대로 이동 
    hanoi(N-1, source, other, destination)
    # N번 원판을 목적 장대로 이동
    ans.append([source, destination])
    # N-1개의 원판을 다시 목적 장대로 이동
    hanoi(N-1, other, destination, source)
    

# 입력
# N: 영상의 크기
N = int(f.readline())
ans= []

hanoi(N, 1, 3, 2)

# 출력 
# 첫째 줄: 옮긴 횟수 K
# 둘째 줄 ~: 수행 과정
print(len(ans))
for i in ans:
    print(*i)
```

##**분할 정복 - 1992번** 쿼드 트리

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

def quad_tree(x, y, size):
    isSameNum = True
    for i in range(x, x+size):
        for j in range(y, y+size):
            if img[x][y] != img[i][j]:
                isSameNum = False
                break
    if not isSameNum:
        ans.append('(')
        for i in range(x, x+size, size//2):
            for j in range(y, y+size, size//2):
                quad_tree(i, j, size//2)
        ans.append(')')
    else:
        ans.append(img[x][y])
        return
    

# 입력
# N: 영상의 크기
N = int(f.readline())
img= [list(map(int, list(f.readline().strip()))) for _ in range(N)]
ans=[]

quad_tree(0, 0, N)

# 출력 
# 영상을 압축한 결과
print(*ans, sep='')
```

##**분할 정복 - 2447번** 별 찍기 - 10

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

def set_blank(x, y, size):
    for i in range(x, x+size):
        for j in range(y, y+size):
            stars[i][j] = ' '

def set_stars(x, y, size):
    if size == 1:
        return
    for i in range(x, x+size, size//3):
        for j in range(y, y+size, size//3):
            if i == x + size//3 and j == y+size//3:
                set_blank(i, j, size//3)
            else:
                set_stars(i, j, size//3)
    

# 입력
# N: 3의 거듭제곱
# N = 3^k, 1 ≤ k < 8
# 패턴: 가운데 공백
N = int(f.readline())
stars=[['*'] * N for _ in range(N)]

set_stars(0, 0, N)


# 출력 
# 첫째 줄부터 N번째 줄까지 별을 출력
for i in stars:
    print(*i, sep='')
```

##**분할 정복 - 2448번** 별 찍기 - 11

```
import sys
import math
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

def set_stars(indent):
    length = len(stars)
    for i in range(length):
        stars.append(stars[i] + stars[i])
        stars[i] = ("   " * indent + stars[i] + "   " * indent)
    

# 입력
# N: 라인 수 (3, 6, 12, 24, 48, ...)
# N = 3*(2^k), 0 ≤ k ≤ 10
N = int(f.readline())
# k-1: 들여쓰기 횟수
k = int(math.log(N//3, 2))
stars = ["  *   ", " * *  ", "***** "]

for i in range(k):
    set_stars(2**i)

# 출력 
# 첫째 줄부터 N번째 줄까지 별을 출력
for i in stars:
    print(i)
```

##**분할 정복 - 1517번** 버블 소트

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

def merge(s, mid, e):
    global count
    left, right = s, mid+1
    result = []
    tmp = 0
    while left <= mid and right <= e:
        if lst[left] < lst[right]:
            result.append(lst[left])
            left += 1
            count += tmp
        else:
            result.append(lst[right])
            right += 1
            tmp += 1
    while left <= mid:
        result.append(lst[left])
        left += 1
        count += tmp
    while right <= e:
        result.append(lst[right])
        right += 1
    
    for i in range(len(result)):
        lst[s+i] = result[i]

def merge_sort(s, e):
    if s >= e:
        return
    mid = (s+e) // 2

    # divide
    merge_sort(s, mid)
    merge_sort(mid+1, e)

    #merge
    merge(s, mid, e)

    
# 입력
# 첫째 줄: N(1≤N≤500,000)
# O(N^2)으로는 풀 수 없다
# 다음 줄: N개의 정수
N = int(f.readline())
lst = list(map(int, f.readline().split()))
count = 0

print(lst)
merge_sort(0, len(lst)-1)

# 출력 
# Swap 횟수
print(count)
```

##**분할 정복 - 2261번** 가장 가까운 두 점

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

# 두 점 사이의 거리
# pow(pos2[0]-pos1[0], 2) + pow(pos2[1]-pos1[1], 2)
def get_distance(pos1, pos2):
    return pow(pos2[0]-pos1[0], 2) + pow(pos2[1]-pos1[1], 2)
    
def get_shortest(s, e):
    dist = e - s + 1
    pivot = (s + e) // 2
    # 두 점이 피봇을 기준으로 왼쪽, 오른쪽 영역에만 속한 경우 
    if dist == 2:
        return get_distance(lst[s], lst[e])
    elif dist == 3:
        return min(get_distance(lst[s], lst[e]), get_distance(lst[s], lst[s+1]), get_distance(lst[e-1], lst[e]))
    else:
        temp = min(get_shortest(s, pivot), get_shortest(pivot + 1, e))
        mid_x = lst[pivot][0]
        candidate = []
        for i in range(s, e+1):
            if pow(mid_x - lst[i][0], 2) <= temp:
                candidate.append(lst[i])
        candidate.sort(key=lambda x: x[1])
        shortest = temp
        # 두 점이 왼쪽, 오른쪽 영역에 모두 걸쳐 있는 경우
        if len(candidate) >= 2:
            for i in range(len(candidate)-1):
                for j in range(i+1, len(candidate)):
                    if pow(candidate[i][1] - candidate[j][1], 2) >= temp:
                        break
                    elif candidate[i][0] < mid_x and candidate[j][0] < mid_x:
                        continue
                    elif candidate[i][0] > mid_x and candidate[j][0] > mid_x:
                        continue
                    else:
                        shortest = min(shortest, get_distance(candidate[i], candidate[j]))
        return shortest
    
# 입력
# 첫째 줄: 자연수 n(2 ≤ n ≤ 100,000)
# 다음 n개의 줄: 각 점의 x, y좌표
# 각각의 좌표는 절댓값이 10,000을 넘지 않는 정수
# 같은 점이 여러 번 주어질 수도 있다
n = int(f.readline())
lst = [list(map(int, f.readline().split())) for _ in range(n)]
lst.sort(key=lambda x: x[0])

print(lst)

# 출력 
# 가장 가까운 두 점의 거리의 제곱
print(get_shortest(0, len(lst)-1))
```

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

# 반복문
def get_min(A, K):
    rest= K
    count = 0
    for i in A[::-1]:
        if rest == 0:
            break
        count += rest // i
        rest %= i
    return count


# 입력
# 첫째 줄: N, K (1 ≤ N ≤ 10, 1 ≤ K ≤ 100,000,000)
# 다음 N개의 줄: 동전의 가치 Ai (오름차순) (1 ≤ Ai ≤ 1,000,000, A1 = 1, i ≥ 2인 경우에 Ai는 Ai-1의 배수)
N, K = map(int, f.readline().split())
A = [int(f.readline()) for _ in range(N)]

# 출력 
# K원을 만드는데 필요한 동전 개수의 최솟값
print(get_min(A, K))
```

##**그리디 - 2875번** 대회 or 인턴

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

# 2명의 여학생과 1명의 남학생이 팀을 결성해서 나가는 것이 원칙
# 대회에 나가는 팀의 수: i
# 대회에 나가는 여학생의 수 : 2*i
# 대회에 나가는 남학생의 수 : i
# 인턴쉽에 참여해야하는 인원 : 
# N - 2*i + M - i >= K
def get_max(N, M, K):
    team = 0
    while N + M >= K and N >= 0 and M >= 0:
        N -= 2
        M -= 1
        team += 1
    return team - 1


# 입력
# 첫째 줄: N, M, K (0 ≤ M ≤ 100, 0 ≤ N ≤ 100, 0 ≤ K ≤ M+N)
# 여학생의 수 N, 남학생의 수 M, 인턴쉽에 참여해야하는 인원 K
N, M, K = map(int, f.readline().split())

# 출력 
# 만들 수 있는 팀의 최대 개수
print(get_max(N, M, K))
```
