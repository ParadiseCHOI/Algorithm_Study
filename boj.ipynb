## **DP - 1463번** 1로 만들기

```
import sys

f = open("input.txt", 'r')
val = int(f.readline())
dp = [0 for _ in range(val+1)]

for i in range(2, val+1):
    dp[i] = dp[i-1] + 1
    if i%3 == 0:
        dp[i] = min(dp[i], dp[i//3] + 1)
    if i%2 == 0:
        dp[i] = min(dp[i], dp[i//2] + 1)

print(dp[val])
```

```
import sys

val = int(input())
# count = int(sys.stdin.readline()))
mod = 1000000000

dp = [[0 for i in range(10)] for j in range(val+1)]

for i in range(1, 10):
  dp[1][i] = 1

for i in range(2, val+1):
  for j in range(10):
    if j == 0:
      dp[i][j] = dp[i-1][j+1]
    elif j == 9:
      dp[i][j] = dp[i-1][j-1]
    else:
      dp[i][j] = dp[i-1][j-1] + dp[i-1][j+1]

print(sum(dp[val])%mod)

```

##**그래프 - 1260번** DFS와 BFS

```
import sys

f = open("input.txt", 'r')
V, E, s = map(int, f.readline().split())

graph = [[0] * (V+1) for _ in range(V+1)]
visited = [False] * (V+1)

for _ in range(E):
    i, j = map(int, f.readline().split())
    graph[i][j] = graph[j][i] = 1

# DFS(Depth-First Search)
# > 재귀를 통해 최대 깊이까지 순차 방문
def dfs(s):
    print(s, end=' ')
    visited[s] = True
    for i in range(1, V+1):
        if (not visited[i]) and (graph[s][i] == 1):
            dfs(i)

# BFS(Breadth-First Search)
# > 큐에 가능한 모든 정점을 넣어둔 후 큐에서 정점을 꺼내가며 방문
def bfs(s):
    queue = [s]
    visited[s] = False
    while queue:
        s = queue.pop(0)
        print(s, end=' ')
        for i in range(1, V+1):
            if (visited[i]) and (graph[s][i] == 1):
                queue.append(i)
                visited[i] = False

dfs(s)
print()
bfs(s)
```

##**그래프 - 11724번** 연결 요소의 개수

```
import sys
sys.setrecursionlimit(10000)

f = open("input.txt", 'r')
V, E = map(int, f.readline().split())

graph = [[0] * (V+1) for _ in range(V+1)]
visited = [False] * (V+1)
count = 0

for _ in range(E):
    i, j = map(int, f.readline().split())
    graph[i][j] = 1

# DFS(Depth-First Search)
# > 재귀를 통해 최대 깊이까지 순차 방문
def dfs(s):
    visited[s] = True
    for i in range(1, V+1):
        if (not visited[i]) and (graph[s][i] == 1):
            dfs(i)

def dfsAll():
    global count
    for i in range(1, V+1):
        if not visited[i]:
            dfs(i)
            count += 1

dfsAll()
print(count)
```

##**분할 정복 - 2261번** 가장 가까운 두 점

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

# 두 점 사이의 거리
# pow(pos2[0]-pos1[0], 2) + pow(pos2[1]-pos1[1], 2)
def get_distance(pos1, pos2):
    return pow(pos2[0]-pos1[0], 2) + pow(pos2[1]-pos1[1], 2)
    
def get_shortest(s, e):
    dist = e - s + 1
    pivot = (s + e) // 2
    # 두 점이 피봇을 기준으로 왼쪽, 오른쪽 영역에만 속한 경우 
    if dist == 2:
        return get_distance(lst[s], lst[e])
    elif dist == 3:
        return min(get_distance(lst[s], lst[e]), get_distance(lst[s], lst[s+1]), get_distance(lst[e-1], lst[e]))
    else:
        temp = min(get_shortest(s, pivot), get_shortest(pivot + 1, e))
        mid_x = lst[pivot][0]
        candidate = []
        for i in range(s, e+1):
            if pow(mid_x - lst[i][0], 2) <= temp:
                candidate.append(lst[i])
        candidate.sort(key=lambda x: x[1])
        shortest = temp
        # 두 점이 왼쪽, 오른쪽 영역에 모두 걸쳐 있는 경우
        if len(candidate) >= 2:
            for i in range(len(candidate)-1):
                for j in range(i+1, len(candidate)):
                    if pow(candidate[i][1] - candidate[j][1], 2) >= temp:
                        break
                    elif candidate[i][0] < mid_x and candidate[j][0] < mid_x:
                        continue
                    elif candidate[i][0] > mid_x and candidate[j][0] > mid_x:
                        continue
                    else:
                        shortest = min(shortest, get_distance(candidate[i], candidate[j]))
        return shortest
    
# 입력
# 첫째 줄: 자연수 n(2 ≤ n ≤ 100,000)
# 다음 n개의 줄: 각 점의 x, y좌표
# 각각의 좌표는 절댓값이 10,000을 넘지 않는 정수
# 같은 점이 여러 번 주어질 수도 있다
n = int(f.readline())
lst = [list(map(int, f.readline().split())) for _ in range(n)]
lst.sort(key=lambda x: x[0])

print(lst)

# 출력 
# 가장 가까운 두 점의 거리의 제곱
print(get_shortest(0, len(lst)-1))
```

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

# 2명의 여학생과 1명의 남학생이 팀을 결성해서 나가는 것이 원칙
# 대회에 나가는 팀의 수: i
# 대회에 나가는 여학생의 수 : 2*i
# 대회에 나가는 남학생의 수 : i
# 인턴쉽에 참여해야하는 인원 : 
# N - 2*i + M - i >= K
def get_max(N, M, K):
    team = 0
    while N + M >= K and N >= 0 and M >= 0:
        N -= 2
        M -= 1
        team += 1
    return team - 1


# 입력
# 첫째 줄: N, M, K (0 ≤ M ≤ 100, 0 ≤ N ≤ 100, 0 ≤ K ≤ M+N)
# 여학생의 수 N, 남학생의 수 M, 인턴쉽에 참여해야하는 인원 K
N, M, K = map(int, f.readline().split())

# 출력 
# 만들 수 있는 팀의 최대 개수
print(get_max(N, M, K))
```

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

def result(E, S, M):
    year = 1
    while True:
        if (year-E) % 15 == 0 and (year-S) % 28 == 0 and (year-M) % 19 == 0:
            print(year)
            break
        year += 1

# 입력
# 첫째 줄: 지구: E, 태양: S, 달: M
# (1 ≤ E ≤ 15, 1 ≤ S ≤ 28, 1 ≤ M ≤ 19)
E, S, M = map(int, f.readline().split())

# 출력 
# 첫째 줄에 E S M으로 표시되는 가장 빠른 연도
result(E, S, M)
```

##**완전탐색 - 1451번** 직사각형으로 나누기

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

def result(N, M, rect):
    # 직사각형을 겹치지 않는 3개의 작은 직사각형으로 나누려 함
    # 작은 직사각형은 그 속에 있는 수의 합
    ans = 0
    rect_sum1, rect_sum2, rect_sum3 = 0, 0 ,0
    # case 1:
    # |||
    # i: 첫번째 사각형의 가로 길이, j: 두번째 사각형의 가로 길이
    for i in range(1, M-1):
        for j in range(i+1, M):
            rect_sum1 = sum([rect[a][b] for a in range(N) for b in range(i)])
            rect_sum2 = sum([rect[a][b] for a in range(N) for b in range(i, j)])
            rect_sum3 = sum([rect[a][b] for a in range(N) for b in range(j, M)])
            ans = max(ans, rect_sum1 * rect_sum2 * rect_sum3)

    # case 2:
    # ≡
    # i: 첫번째 사각형의 세로 길이, j: 두번째 사각형의 세로 길이
    for i in range(1, N-1):
        for j in range(i+1, N):
            rect_sum1 = sum([rect[a][b] for a in range(i) for b in range(M)])
            rect_sum2 = sum([rect[a][b] for a in range(i, j) for b in range(M)])
            rect_sum3 = sum([rect[a][b] for a in range(j, N) for b in range(M)])
            ans = max(ans, rect_sum1 * rect_sum2 * rect_sum3)

    # case 3:
    # ─
    # ||
    # 아래 두 사각형: i: 첫번째 사각형의 세로 길이, j: 두번째 사각형의 세로 길이
    for i in range(1, N):
        for j in range(1, M):
            rect_sum1 = sum([rect[a][b] for a in range(i) for b in range(M)])
            rect_sum2 = sum([rect[a][b] for a in range(i, N) for b in range(j)])
            rect_sum3 = sum([rect[a][b] for a in range(i, N) for b in range(j, M)])
            ans = max(ans, rect_sum1 * rect_sum2 * rect_sum3)

    # case 4:
    # ||
    # ─
    # 위 두 사각형: i: 첫번째 사각형의 세로 길이, j: 두번째 사각형의 세로 길이
    for i in range(1, N):
        for j in range(1, M):
            rect_sum1 = sum([rect[a][b] for a in range(i) for b in range(j)])
            rect_sum2 = sum([rect[a][b] for a in range(i) for b in range(j, M)])
            rect_sum3 = sum([rect[a][b] for a in range(i, N) for b in range(M)])
            ans = max(ans, rect_sum1 * rect_sum2 * rect_sum3)
    # case 5:
    # |=
    for i in range(1, N):
        for j in range(1, M):
            rect_sum1 = sum([rect[a][b] for a in range(N) for b in range(j)])
            rect_sum2 = sum([rect[a][b] for a in range(i) for b in range(j, M)])
            rect_sum3 = sum([rect[a][b] for a in range(i, N) for b in range(j, M)])
            ans = max(ans, rect_sum1 * rect_sum2 * rect_sum3)

    # case 6:
    # =|
    for i in range(1, N):
        for j in range(1, M):
            rect_sum1 = sum([rect[a][b] for a in range(i) for b in range(j)])
            rect_sum2 = sum([rect[a][b] for a in range(i, N) for b in range(j)])
            rect_sum3 = sum([rect[a][b] for a in range(N) for b in range(j, M)])
            ans = max(ans, rect_sum1 * rect_sum2 * rect_sum3)
    print(ans)
                

# 입력
# 첫째 줄: 세로: N, 가로: M (1 <= N, M <= 100) (N >=3 || M >=3)
# 둘째 줄 ~ N번째 줄: 직사각형에 들어가는 수 (직사각형에는 적어도 3개의 수)
# (각 줄에 M개의 수, 직사각형에 들어가는 수는 한 자리의 숫자)
N, M = map(int, f.readline().split())
rect = [list(map(int, list(f.readline().strip()))) for _ in range(N)]

# 출력 
# 세 개의 작은 직사각형의 합의 곱의 최댓값
result(N, M, rect)
```

##**완전탐색 - 1963번** 소수 경로

```
import sys
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

def isPrime(num):
    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            return False
    return True

def solve(case):
    q = deque([[case[0], 0]])
    while q:
        curr, count = q.popleft()
        if curr == case[1]:
            print(count)
            return

        str_curr = str(curr)
    
        for i in range(4):
            for j in map(str, range(10)):
                # 가장 앞 자리수가 0인 경우 제외
                if i == 0 and j =='0':
                    continue
                num = int(str_curr[:i] + j + str_curr[i+1:])
                if prime[num] and visited[num] == 0:
                    visited[num] = 1
                    q.append([num, count+1])
            


# 입력
# 첫째 줄: test case의 수 T
# T개의 줄: 각 줄에 1쌍씩 네 자리 소수 (항상 네 자리 소수)
T = int(f.readline())
cases = [list(map(int, f.readline().split())) for _ in range(T)]
prime = [False for _ in range(10001)]

for i in range(1000, 10001):
    prime[i] = isPrime(i)

# 출력 
# 각 test case에 대해 두 소수 사이의 변환에 필요한 최소 회수
# 불가능한 경우 Impossible을 출력
for i in cases:
    visited = [0]*10001
    solve(i)
```

##**완전탐색 - 9019번** DSLR

```
import sys
from collections import deque
sys.setrecursionlimit(1000000)

f = open("input.txt", 'r')

# PyPy3로 제출했을 경우만 통과
# Python3 시간초과
def solve(case):
    global mod
    # n의 네 자릿수를 d1, d2, d3, d4
    # (n = ((d1 × 10 + d2) × 10 + d3) × 10 + d4)
    q = deque([[case[0], '']])
    while q:
        curr, ans= q.popleft()
        if curr == case[1]:
            print(ans)
            return

        # case D:
        # 1. D: D 는 n을 두 배로 바꾼다.
        # - 결과 값이 9999 보다 큰 경우에는 10000 으로 나눈 나머지
        # - 결과 값(2n mod 10000)을 레지스터에 저장
        tmp = (curr * 2) % mod
        if visited[tmp] == 0:
            visited[tmp] = 1
            q.append([tmp, ans+'D'])
        # case S:
        # 2. S: S 는 n에서 1 을 뺀 결과 n-1을 레지스터에 저장
        # - n이 0 이라면 9999 가 대신 레지스터에 저장
        tmp = (curr + 9999) % mod
        if visited[tmp] == 0:
            visited[tmp] = 1
            q.append([tmp, ans+'S'])
        # case L:
        # 3. L: L 은 n의 각 자릿수를 왼편으로 회전시켜 그 결과를 레지스터에 저장
        # - 연산이 끝나면 레지스터에 저장된 네 자릿수는 왼편부터 d2, d3, d4, d1
        tmp = (curr*10) % mod + (curr//1000)
        if visited[tmp] == 0:
            visited[tmp] = 1
            q.append([tmp, ans+'L'])
        # case R:
        # 4. R: R 은 n의 각 자릿수를 오른편으로 회전시켜 그 결과를 레지스터에 저장
        # - 연산이 끝나면 레지스터에 저장된 네 자릿수는 왼편부터 d4, d1, d2, d3
        tmp = curr // 10 + (curr % 10) * 1000
        if visited[tmp] == 0:
            visited[tmp] = 1
            q.append([tmp, ans+'R'])
            


# 입력
# 첫째 줄: test case의 수 T
# T개의 줄: 두 개의 정수 A와 B (A ≠ B, 0 <= A, B < 10,000)
# A: 레지스터의 초기 값, B: 최종 값
T = int(f.readline())
cases = [list(map(int, f.readline().split())) for _ in range(T)]
mod = 10000

# 출력 
# A에서 B로 변환하기 위해 필요한 최소한의 명령어 나열
for i in cases:
    visited = [0]*10000
    solve(i)
```

##**완전탐색 - 2251번** 물통

```
import sys
from collections import deque

f = open("input.txt", "r")
def solve(A, B, C):
    # 처음에는 앞의 두 물통은 비어 있고
    # 세 번째 물통은 가득(C 리터) 차 있다.
    # 한 물통이 비거나, 다른 한 물통이 가득 찰 때까지 물을 부을 수 있다
    q = deque([[0, 0]])
    ans = dict()
    while q:
        na, nb = q.popleft()
        nc = C - na - nb
        # 첫 번째 물통이 비어 있을 때,
        # 세 번째 물통(용량이 C인)에 담겨있을 수 있는 물의 양
        if na == 0:
            ans[nc] = 1
        
        # C에서 다른 물통
        if nc > 0:
            # C >> A
            if na < A:
                water = min(nc, A - na)
                if visited[na + water][nb] == 0:
                    visited[na + water][nb] = 1
                    q.append([na + water, nb])
            # C >> B
            if nb < B:
                water = min(nc, B - nb)
                if visited[na][nb + water] == 0:
                    visited[na][nb + water] = 1
                    q.append([na, nb + water])

        # B에서 다른 물통
        if nb > 0:
            # B >> A
            if na < A:
                water = min(nb, A - na)
                if visited[na + water][nb - water] == 0:
                    visited[na + water][nb - water] = 1
                    q.append([na + water, nb - water])
            # B >> C
            if nc < C:
                water = min(nb, C - nc)
                if visited[na][nb - water] == 0:
                    visited[na][nb - water] = 1
                    q.append([na, nb - water])

        # A에서 다른 물통
        if na > 0:
            # A >> B
            if nb < B:
                water = min(na, B - nb)
                if visited[na - water][nb + water] == 0:
                    visited[na - water][nb + water] = 1
                    q.append([na - water, nb + water])
            # A >> C
            if nc < C:
                water = min(na, C - nc)
                if visited[na - water][nb] == 0:
                    visited[na - water][nb] = 1
                    q.append([na - water, nb])

    for i in range(C+1):
        if ans.get(i):
            print(i, end =  " ")

   
# 입력
# 첫째 줄: 세 정수 A, B, C (1≤A, B, C≤200)
A, B, C = map(int, f.readline().split())

# 출력
# 첫 번째 물통(용량이 A인)이 비어 있을 때,
# 세 번째 물통(용량이 C인)에 담겨있을 수 있는 모든 물의 양
# 공백으로 구분하여 답을 출력, 각 용량은 오름차순으로 정렬
visited = [[0] * (B+1) for _ in range(A+1)]
solve(A, B, C)
```
