#이진수 구하기
1. 반복문
2. 재귀함수

```
# 1. 반복문
x = int(input())
result = ''
while True:
  if x == 1:
    result += str(x)
    print(result[::-1])
    break
  if x % 2 == 0:
    result += '0'
  else:
    result += '1'
  x = x // 2
```

```
# 2. 재귀함수
result = ''
def getBin(num):
  if num == 1:
    return str(num)
  else:
    return getBin(num//2) + str(num%2)

print(getBin(10))
```

```
getFibonachi = lambda x: 1 if x == 1 or x == 0 else getFibonachi(x-1) + getFibonachi(x-2)
getFibonachi(5)
```
